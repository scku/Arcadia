```python
%%R

data_housing <- read.csv("trulia_data.csv", stringsAsFactors=FALSE)
cat("Housing data summary:"); print(summary(data_housing))
```

```python
%%R -w 600

library(ggplot2)

data_housing$date <- as.Date(data_housing$weekEndingDate, format="%Y-%m-%d")
print(ggplot(data=data_housing, aes(date, allNum)) + geom_point(colour="blue") + geom_line() + ylim(0,300) + labs(x="Date", y="Count", title="Number of House Listings"))
print(ggplot(data=data_housing, aes(date, allAvg)) + geom_point(colour="blue") + geom_line() + labs(x="Date", y="Price", title="Average Price of House Listings"))
```
