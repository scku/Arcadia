## Housing Data
I utilized the Trulia API to gather all the available home listings in Arcadia, CA. I wanted to gather all the data from the past decade (2003 to 2013) but the data only goes back to mid-2009. See the separate Python script in the repository for the API call and data cleaning.

```python
%load_ext rmagic
import rpy2 as Rpy
```

```python
%%R

library(ggplot2)
library(reshape)
library(plotrix)
```

```python
%%R

data_housing <- read.csv("trulia_data.csv", stringsAsFactors=FALSE)
cat("Housing data summary:"); print(summary(data_housing))
```

```python
%%R -w 1000

data_housing$date <- as.Date(data_housing$weekEndingDate, format="%Y-%m-%d")
print(ggplot(data=data_housing, aes(date, allNum)) + geom_point(colour="blue") + geom_line() + ylim(0,300) + labs(x="Date", y="Count", title="Number of House Listings"))
print(ggplot(data=data_housing, aes(date, allAvg)) + geom_point(colour="blue") + geom_line() + labs(x="Date", y="Price", title="Average Price of House Listings"))

data_housing.long <- melt(data_housing, id=c("date"), measure=c("X2Num", "X3Num", "X4Num", "X5Num", "X6Num", "X7Num", "X8Num"))
print(ggplot(data=data_housing.long, aes(date, value, colour=variable)) + geom_line() + labs(x="Date", y="Count", title="Number of House Listings By Number of Bedrooms"))
```

The average price of house listings is very surpising. The average price went from around \$1 million in the start of 2012 to almost \$2 million in early 2014. Within the span of 2 years, the average price increased by close to \$1 million. Looking at the number of bedrooms, there is a decrease of house listings with 2-3 bedrooms while the higher bedroom count houses are relatively flat. From my personal observation, there is definitely a trend of buying up older homes with large plots of land, demolishing, and building new mansion-style houses. There is a few of such examples within a few blocks of my house that have recently sold for around \$4 million. 

#Demographics
Let's look at the 2000 Census, 2010 Census, and 2012 American Community Survey.

```python
%%R -w 1000

census2000_data <- read.csv("2000census.csv", stringsAsFactors=FALSE)
census2010_data <- read.csv("2010census.csv", stringsAsFactors=FALSE)
acs2012_data <- read.csv("2012acs.csv", stringsAsFactors=FALSE)

census2000_data.one_race <- census2000_data[,54:89]
census2000_data.one_race.white <- strtoi(census2000_data.one_race[2,3])
census2000_data.one_race.black <- strtoi(census2000_data.one_race[2,5])
census2000_data.one_race.indian <- strtoi(census2000_data.one_race[2,7])
census2000_data.one_race.asian <- strtoi(census2000_data.one_race[2,9])
census2000_data.one_race.pacific <- strtoi(census2000_data.one_race[2,25])

census2010_data.one_race.white <- strtoi(census2010_data[2,158])
census2010_data.one_race.black <- strtoi(census2010_data[2,160])
census2010_data.one_race.indian <- strtoi(census2010_data[2,162])
census2010_data.one_race.asian <- strtoi(census2010_data[2,164])
census2010_data.one_race.pacific <- strtoi(census2010_data[2,180])

#print(census2000_data.one_race.white); print(census2000_data.one_race.black)
#print(census2000_data.one_race.indian); print(census2000_data.one_race.asian)
#print(census2000_data.one_race.pacific)

#print(census2010_data.one_race.white); print(census2010_data.one_race.black)
#print(census2010_data.one_race.indian); print(census2010_data.one_race.asian)
#print(census2010_data.one_race.pacific)

#par(mfrow=c(2,2))
slices <- c(census2000_data.one_race.white, census2000_data.one_race.black, census2000_data.one_race.indian,
            census2000_data.one_race.asian, census2000_data.one_race.pacific)
lbls <- c("White", "Black or African American", "American Indian and Alaska Native", 
          "Asian", "Native Hawaiian and Other Pacific Islander")
pct <- round(slices/sum(slices)*100)
lbls <- paste(lbls, pct) # add percents to labels 
lbls <- paste(lbls,"%",sep="") # add % to labels 
pie(slices, labels=lbls, col=rainbow(length(lbls)), main="Pie Chart of Demographics from 2000 Census")

slices <- c(census2010_data.one_race.white, census2010_data.one_race.black, census2010_data.one_race.indian,
            census2010_data.one_race.asian, census2010_data.one_race.pacific)
lbls <- c("White", "Black or African American", "American Indian and Alaska Native", 
          "Asian", "Native Hawaiian and Other Pacific Islander")
pct <- round(slices/sum(slices)*100)
lbls <- paste(lbls, pct) # add percents to labels 
lbls <- paste(lbls,"%",sep="") # add % to labels 
pie(slices, labels=lbls, col=rainbow(length(lbls)), main="Pie Chart of Demographics from 2010 Census")
```
